```python
%matplotlib qt
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
import random
import numpy as np

# Initialisation
x = []
y = []
colors = []
fig, ax = plt.subplots(figsize=(7,5)) # On définit 'ax' pour plus de stabilité

def animation_func(i):
    # On ajoute les données
    x.append(random.randint(0,100))
    y.append(random.randint(0,100))
    colors.append(np.random.rand(3)) # rand(3) pour une couleur RGB valide
    area = random.randint(10, 500)   # Taille plus visible
    
    # Au lieu de plt.scatter, on nettoie et on redessine sur l'axe
    ax.clear() 
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.scatter(x, y, c=colors, s=area, alpha=0.5)

# Correction des warnings : 
# 1. cache_frame_data=False évite la saturation de la mémoire
# 2. On s'assure que 'animation' est bien affecté
animation = FuncAnimation(fig, animation_func, 
                          interval=100, 
                          cache_frame_data=False,
                          save_count=100)

plt.show()
```
